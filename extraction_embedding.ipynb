```python
import os
import pandas as pd
import numpy as np
import soundfile as sf
import openl3
```

```python
def extract_features(audio_path: str):
    audio, sr = sf.read(audio_path)
    embedding, ts = openl3.get_audio_embedding(audio, sr, input_repr="mel256", embedding_size=512)
    return embedding

def edit_path(path: str):
    dir_name, file_name = os.path.split(path)
    return f"./data_with_wav/wav_clips/{file_name}"

def get_embedding(path: str):
    if os.path.exists(path):
        return extract_features(path)
    else:
        return None
```

```python
# Reading Data
openl3_data = pd.read_csv(f'data_with_wav/data.csv')
openl3_data.head()
```

```python
openl3_data['wav Path'] = openl3_data['wav Path'].apply(edit_path)
embeddings_df = pd.DataFrame(columns=['npy_path', 'emotion'])
```

```python
counter = 0
total = len(openl3_data)
for index, row in openl3_data.iterrows():
    wav_path = row['wav Path']
    emotion = row['Emotion']
    print(f"{counter} from {total}")
    if os.path.exists(wav_path):

        # Get the embedding
        embedding = get_embedding(wav_path)

        file_name = os.path.basename(wav_path).replace('.wav', '.npy')
        npy_dir = os.path.join(os.path.dirname(os.path.dirname(wav_path)), 'openl3_embedding')
        npy_path = os.path.join(npy_dir, file_name)

        os.makedirs(npy_dir, exist_ok=True)
        np.save(npy_path, embedding)
        embeddings_df = embeddings_df.append({'npy_path': npy_path, 'emotion': emotion}, ignore_index=True)
        
    counter += 1

# Save embeddings into new csv
embeddings_df.to_csv('embeddings_data.csv', index=False)
```
